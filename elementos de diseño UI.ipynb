### Consejos colores
* Un color resalta más si es menos utilizado y cuantos menos colores estén en la pantalla, no sobreutilizar un color
* Comenzar a diseñar en escala de grises, crear una interfaz solida
* Jugar con el color
* Resaltar acciones importantes
* Crear jerarquía visual
* utilizar paletas cromáticas
#### Regla 60-30-10
Se pueden ocupar variaciones de los tres colores ya sea mas oscuras o más claras
* Los colores ganan armonía cuando conviven 3 colores diferentes
* Los colores deben estar presentes en una proporción 60 30 10<br>
<br>En el 60% de los casos se utilizara un color
<br>El 30% al segundo color, color complementario
<br>10% El color con más peso visual pero que menos se ocupa, elementos que debén ser resaltados 

#### Contraste
Relación entre el color del objeto y el fondo<br>
Utilizar color review, nos dirá los ratios de contraste
<br>Intalar plugin A11 color contrast checker
<br>
<br>
<b>AA</b><br>
El Mayor contraste se obtiene de 21:1 cuando el fondo es negro y el elemento es blanco o alrevés
<br> El mínimo ratio de contraste es de 4,5:1 en texto normal
<br> 3:1 para textos grandes, o componentes de interfaz, como bordes de inputs de formularios
<br>
<b>AAA</b>
<br>Ratio de 7:1 para texto normal
<br>4,5:1 para los demás componentes

#### Jerarquía visual
Determinar la jerarquía visual, partiendo de una escala de grises, se pinta el elemento que queremos resaltar luego vamos al segundo elemento y vemos que tanto peso visual tiene para que no le quite peso visual al elemento que más queremos resaltar, si ya tiene mucho tamaño poner un color suave

* Evitar tipografías que tienen letras casi similares

* Tener un peso distinto(grosor) pero no demasiado
* Tener un estilo similar

Estilos o tipos de letras que podemos elegir<br>

Son diferentes fuentes, pero ultimadamente solo tienen una fuente lo que se puede confundir tipografia con fuente<br>
<b>tipo</b>(forma)<b> grafia</b>(escritura)

## Tipografías

<b>Serif</b> tiene una sariza al final de las letras, facilitan la lectura, se usan en libros escritos<br>
<p>ejemplo:
Georgia<br>
Times New Roman<br>
<br>
<b>Sans-serif</b> el final de las letras es cuadrado, se usan en contextos digitales, menos cantidad de texto<br>
<p>ejemplo:
Arial<br>
Helvetica<br>

## Fuente
Variación de peso, estilo, tamaño, y efecto de una tipografía
<p>ejemplo:
Regular<br>
Bold<br>
Italic<br>
Bold italic<br>

## Anatomía de la letra
<b>Altura de la X</b>
Se refiere a cuanto miden las letras mínusculas<br>
<b> Línea base</b>
Linea en la que se apoyan las letras<br>
<b> Asta ascendiente</b>
Se refiere a las letras que sobresalen de la altura de la x<br>
<b> Asta descendiente</b>
La parte de las letras que sobresalen en la parte de abajo de la base de la X<br>

### Longitud de línea
Cuando la línea es muy larga el ojo del lector, tendrá dificultad para continuar en la línea correcta, pero, si es demasiado corta el lector tendrá un corte de lectura estresando al lector<br>
<br>Nuestros texto no debe adaptarse al diseño por el contrario debemos adaptar nuestros diseños 

<b>Consejos de texto:</b>
<p>Tamaño

*  Debe contener entre 45 y 75 espacios de puntuación
<p style="padding-left: 40px">Método de bringhurts: tamaño de la letra x 30

* Las letras no deben ser ni tan grandes ni tan pequeñas
* Deben adaptarse al contexto de uso
* Evitar tamaño de fuentes menores a 12 pixeles</p>
<p>Interlineado

* Debe ser de 150% del tamaño de la fuente, se debe incrementar en dispositivos móbiles
</p>
<p>Alineación

* Utilizar texto siempre hacía la izquierda
* Texto centrado solo en espacios muy reducidos o cuando es muy poco texto
* Evitar el texto justificado
</p>

# Iconos
* Ahorran espacio
* Deben tener simplicidad
* Acompañar con texto(opcional) 
* Deben ser entendidos por si solos
* Iconos con máximo de dos colores
* No muy detallados 
* Posicionarlos encima de un color solido
* Visible y compresible
* Lo más universal posible
* Unidad de estilo, mantener una armonía
<p style="padding-left: 60px">Debén compartir la misma paleta cromática<br>Mismo peso<br>Mismo ancho de linea<be>Mismo nivel de detalle<p style="padding-left: 50px">

* Unicidad
<p style="padding-left: 60px">Los icónos que sean originales serán mejor recordados<p style="padding-left: 50px">

* Si utilizamos iconos grandes debemos elegir icónos con más detalles

## Enlaces
Las personas no desperdician clicks
* No hacer dudar de que es un enlace
* Deben llamar la atencion
* Describir con presición la página de destino
* Debe responder la pregunta ¿Qué obtendrá el usuario cuando haga click aquí?
* Texto único
* Empezar con palabras clave 
* Aspecto visual
* Debe sugerir ser clicables
* Asignar un color a los enlaces
* Debe funcionar para personas daltonicas
* No subrayar en menús
